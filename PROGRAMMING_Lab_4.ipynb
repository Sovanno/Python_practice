```python
import pandas as pd
```

```python
df_c = pd.read_csv('..\data\weatherAUS.csv')
```

```python
df_c.info()
```

```python
missing_values = df_c.isnull().sum()
missing_values
```

```python
for column in df_c.columns:
    if df_c[column].dtype == 'int64':
        df_c[column].fillna(df_c[column].median(), inplace=True)
    elif df_c[column].dtype == 'float64':
        df_c[column].fillna(df_c[column].mean(), inplace=True)
    else:
        df_c[column].fillna(df_c[column].mode()[0], inplace=True)
```

```python
df_c = df_c.rename (
    columns = {
        'Date' : 'date',
        'Location' : 'location',
        'MinTemp' : 'min_temp',
        'MaxTemp' : 'max_temp',
        'Rainfall' : 'rainfall',
        'Evaporation' : 'evaporation',
        'Sunshine' : 'sunshine',
        'WindGustDir' : 'wind_gust_dir',
        'WindGustSpeed' : 'wind_gust_speed',
        'WindDir9am' : 'wind_dir_9am',
        'WindDir3pm' : 'wind_dir_3pm',
        'WindSpeed9am' : 'wind_speed_9am',
        'WindSpeed3pm' : 'wind_speed_3pm',
        'Humidity9am' : 'humidity_9am',
        'Humidity3pm' : 'humidity_3pm',
        'Pressure9am' : 'pressure_9am',
        'Pressure3pm' : 'pressure_3pm',
        'Cloud9am' : 'cloud_9am',
        'Cloud3pm' : 'cloud_3pm',
        'Temp9am' : 'temp_9am',
        'Temp3pm' : 'temp_3pm',
        'RainToday' : 'rain_today',
        'RainTomorrow' : 'rain_tomorrow'
    }
)
```

```python
df_c = df_c.drop('date', axis=1)
df_c
```

```python
df_c.duplicated().sum()
```

```python
df_c = df_c.drop_duplicates().reset_index(drop=True)
```

```python
df_c = df_c.drop('location', axis=1)
```

```python
df_c['rain_today'] = pd.Categorical(df_c['rain_today']).codes
df_c['rain_tomorrow'] = pd.Categorical(df_c['rain_tomorrow']).codes
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
columns_to_encode = ['wind_gust_dir', 'wind_dir_9am', 'wind_dir_3pm']

encoder = OneHotEncoder(sparse=False)
data_encoded = encoder.fit_transform(df_c[columns_to_encode])

encoded_columns = encoder.get_feature_names_out(columns_to_encode)
data_encoded_df = pd.DataFrame(data_encoded, columns=encoded_columns)
data_final = pd.concat([df_c, data_encoded_df], axis=1)

df_c = data_final.drop(columns_to_encode, axis=1)
```

```python
columns_to_process = df_c.columns
```

```python
columns_to_process = columns_to_process.drop("rain_tomorrow")
```

```python
for column in columns_to_process:
    q1 = df_c[column].quantile(0.25)
    q3 = df_c[column].quantile(0.75)
    iqr = q3 - q1
    lower_threshold = q1 - 1.5 * iqr
    upper_threshold = q3 + 1.5 * iqr
    df_c = df_c.loc[(df_c[column] >= lower_threshold) & (df_c[column] <= upper_threshold)]
```

```python
y_c = df_c["rain_tomorrow"]
X_c = df_c.drop(["rain_tomorrow"], axis=1)

y_cr = df_c["max_temp"]
X_r = df_c.drop(["max_temp"], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
normalizer = StandardScaler()
X_normalized = normalizer.fit_transform(X_c)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_c_train, y_c_test = train_test_split(X_normalized, y_c, test_size=0.2, random_state=42)
X_train_r, X_test_r, y_c_train_r, y_c_test_r = train_test_split(X_r, y_cr, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.utils.validation import check_X_y, check_array, check_is_fitted
from sklearn.utils.multiclass import check_classification_targets
```

```python
class BaggingClassifier:
    def __init__(self, base_classifier, n_estimators):
        self.base_classifier = base_classifier
        self.n_estimators = n_estimators
        self.classifiers = []
         
    def fit(self, X, y):
          for _ in range(self.n_estimators):
              indices = np.random.choice(len(X), len(X), replace=True)
              X_sampled = X[indices]
              y_sampled = y[indices]
 
              classifier = self.base_classifier.__class__()
              classifier.fit(X_sampled, y_sampled)
 
              self.classifiers.append(classifier)
          return self.classifiers
    def predict(self, X):
        predictions = [classifier.predict(X) for classifier in self.classifiers]
        majority_votes = np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=0, arr=predictions)
 
        return majority_votes
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
bag_clf = BaggingClassifier(base_classifier=DecisionTreeClassifier(),
                           n_estimators=100)
bag_clf.fit(np.array(X_train), np.array(y_c_train))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
y_pred = bag_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_c_test, np.round(y_pred))
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.base import clone

class BaggingRegressor:
    def __init__(self, base_regressor, n_estimators):
        self.base_regressor = base_regressor
        self.n_estimators = n_estimators
        self.regressors = []

    def fit(self, X, y):
        for _ in range(self.n_estimators):
            indices = np.random.choice(len(X), len(X), replace=True)
            X_sampled = X[indices]
            y_sampled = y[indices]

            regressor = clone(self.base_regressor)
            regressor.fit(X_sampled, y_sampled)

            self.regressors.append(regressor)
        return self

    def predict(self, X):
        predictions = [regressor.predict(X) for regressor in self.regressors]

        ensemble_prediction = np.mean(predictions, axis=0)

        return ensemble_prediction
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
bag_clf = BaggingRegressor(base_regressor=DecisionTreeRegressor(),
                           n_estimators=100)
bag_clf.fit(np.array(X_train_r), np.array(y_c_train_r))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_pred = bag_clf.predict(X_test_r)
```

```python
print(f'MAE: {mean_absolute_error(y_c_test_r, y_pred)}')
```
