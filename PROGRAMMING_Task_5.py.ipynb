```python
import pandas as pd
```

# 1)Создаём датафрейм с данными которые предоставили

```python
data = {'Name': ['Strom, Mrs. Wilhelm (Elna Matilda Persson)', 'Navratil, Mr. Michel ("Louis M Hoffman")',
                     'Minahan, Miss. Daisy E'],
            'Age': [29, 36.5, 33],
            'Sex': ['female', 'male', 'female']}

df = pd.DataFrame(data)
df
```

# 2)Загржаем данные из файла

```python
df_titanic = pd.read_csv('../data/titanic_csv.csv', sep=";")
df_titanic
```

# 3)Загружаем напрямую данные по ссылке:

```python
gist_df = pd.read_csv('https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt')
gist_df
```

# 4)Объединяем датафреймы из п. 2, 3 в один датафрейм, удаляем дубликаты

```python
df_titanic.rename(str.lower, axis='columns', inplace=True)

# Объединение датафреймов
df_concat = pd.concat([df_titanic, gist_df], ignore_index=True)

# Удаление дубликатов строк
df_concat = df_concat.drop_duplicates()

df_concat
```

# 5)Зададим в качестве индекса датафрейма passengerid. Отсортируем по индексу (sort_values).

```python
df_concat = df_concat.set_index('passengerid')  # Задаем индекс
df_concat = df_concat.sort_values(by='passengerid')  # Сортируем по индексу
df_concat
```

# 7)Поменяем местами нулевую по порядку строку с элементом с индексом 2 (loc, iloc).

```python
temp = df_concat.iloc[0].copy()
df_concat.iloc[0] = df_concat.loc[2]
df_concat.iloc[2] = temp
df_concat
```

# 8)В столбце Sex заменяем female/male на f/m (map).

```python
# Создание словаря для замены значений
replacement = {'female': 'f', 'male': 'm'}

# Замена значений с помощью map
df_concat['sex'] = df_concat['sex'].map(replacement)
```

# 9)Найдём билеты по которым плыло 6 и более человек, отобразим список людей).

```python
# Фильтрование билетов с помощью группировки и подсчета
filtered_tickets = df_concat.groupby(['ticket'])['ticket'].count()

# Вывод списка людей на выбранных билетах
#filtered_tickets.name.unique()
filtered_tickets[filtered_tickets.values >= 6]
```
