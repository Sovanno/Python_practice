# 6)Проверим наличие пропусков в данных. Если в каких либо столбцах присутствуют пропуски - заполним их в соответствии со следующим правилом:

```python
missing_values = df.isnull().sum()
missing_values
```

```python
for column in df.columns:
    if df[column].dtype == 'int64':
        df[column].fillna(df[column].median(), inplace=True)
    elif df[column].dtype == 'float64':
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
missing_values = df.isnull().sum()
missing_values
```

# 7) Выберем произвольный числовой признак, с помощью среза данных, выберем не более 200 значений соответствующего признака. Проверим полученную выборку с помощью normaltest на нормальность распределения.

```python
from scipy import stats
```

```python
# Выбор произвольного числового признака
numeric_feature = df['price_card']
sample = numeric_feature.sample(200)
# Проверка выборки на нормальность с помощью теста normaltest
statistic, p_value = stats.normaltest(sample)
# Вывод результатов теста
print(statistic)
print(p_value)
if p_value < 0.05:
    print("Распределение не является нормальным")
else:
    print("Распределение является нормальным")
```
